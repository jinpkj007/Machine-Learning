```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots_adjust 

import numpy as np
f = np.loadtxt('nonParamTrain.txt')
subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=None, hspace=1.8)
plt.subplot(3,1,1)
binwidth=0.2
plt.hist(f,bins=np.arange(min(f), max(f) + binwidth, binwidth));
plt.title('Histogram Density for binwidth =%f' %binwidth)
plt.ylabel('count')
plt.xlabel('bins')
binwidth=0.5
plt.subplot(3,1,2)
plt.hist(f,bins=np.arange(min(f), max(f) + binwidth, binwidth));
plt.title('Histogram Density for binwidth =%f' %binwidth)
plt.ylabel('count')
plt.xlabel('bins')
binwidth=2.0
plt.subplot(3,1,3)
plt.hist(f,bins=np.arange(min(f), max(f) + binwidth, binwidth));
plt.title('Histogram Density for binwidth =%f' %binwidth)
plt.ylabel('count')
plt.xlabel('bins')
```
